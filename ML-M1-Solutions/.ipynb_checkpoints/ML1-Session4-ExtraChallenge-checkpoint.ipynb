### 3. Deploy model through a GUI

```python
def model_correct(button):
    model.count_correct = model.count_correct + 1
    add_to_budget(10)
```

```python
def model_wrong(buttom):
    model.count_wrong = model.count_wrong + 1
    remove_from_budget(10)
```

```python
main = start_layout("30%", "70%", "20%")
title = create_title_text(description = "Clothes Classifier", size = "30pt", height = "80%")
upload2 = create_fileUploadButton(description = "Load Image", height = "90%", file_type = "Image")
button6 = create_button(description = 'Predict Image', height = "90%", button_color = "white")
button_convert = create_button(description = "Convert to Negative Image", height = "90%")
link_button(button_convert, convert_to_negative, main)
link_button(button6, show_prediction_results, main,  model)
text = create_normal_text(color = "blue",description = "This application allows users to upload their own image to predict the type of clothes")
feedback_yes = create_button(description = 'Yes', height = "30%", button_color = "Green")
feedback_no = create_button(description = 'No', height = "30%", button_color = "Red")
feedback_yes.on_click(model_correct)
feedback_no.on_click(model_wrong)
place_elements(main, text, location = "footer")
place_elements(main, title, location = "header")
place_elements(main, upload2,button_convert, button6, location = "header", alignment = "Horizontal")
place_elements(main, feedback_yes,feedback_no,  location = "footer", alignment = "Horizontal")

display_main(main)
```
